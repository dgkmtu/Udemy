```python
from flickrapi import FlickrAPI
from urllib.request import urlretrieve
from pprint import pprint
from tqdm import tqdm
import os, time, sys
```

```python
# APIキーの情報
key = "4790622ca19ec3c6aac84ffce4ce8a95"
secret = "a108c90fa6953a51"
wait_time = 1
```

```python
# fish_list.csvを開き、スクレイピング対象を取得する → serch_list
search_list = []
with open('fish_list.csv', mode='r', encoding='utf-8') as f:
    for row in f:
        column = row.rstrip()
        search_list.append(column)
print(search_list)
```

```python
for fishname in search_list:
    # 保存フォルダの作成
    os.mkdir(fishname)
```

```python
for fishname in search_list:
    # 保存フォルダの指定
    savedir = './' + fishname
    
    # FlickrAPIの設定
    flickr = FlickrAPI(key, secret, format='parsed-json')
    result = flickr.photos.search(
        text = fishname,
        per_page = 600,
        media = 'photos',
        safe_search = 1,
        extras = 'url_q, licence'
    )
    
    # スクレイピングの実行
    photos = result['photos']
    
    for i, photo in tqdm(enumerate(photos['photo'])):
        url_q = photo['url_q']
        filepath = savedir + '/' + photo['id'] + '.jpg'
        if os.path.exists(filepath): continue
        urlretrieve(url_q, filepath)
        time.sleep(wait_time)
```
